```python
#Import data set

import pandas as pd

train = pd.read_csv("train.csv")
train
```

```python
#Indetify Variables
train.columns
```

```python
train.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
#Histogram of Sales price
plt.hist(train['SalePrice'], bins=30)
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.title('Histogram of Sale Price')
plt.show()
```

```python
#Data Set statistics
train.describe()
```

```python
#Correlation coefficient between Sales price and Lot area
price_are = train['SalePrice'].corr(train['LotArea'])

print(price_are)
```

```python
import matplotlib.pyplot as plt


x_lot = train['LotArea']
y_price = train['SalePrice'] 

# Create the scatter plot
plt.scatter(x_lot, y_price, alpha=0.5)

# Add axis labels and a title
plt.xlabel('Lot Area')
plt.ylabel('Sale Price')
plt.title('Lot Area vs. Sales Price Scatter Plot')

# Display the plot
plt.show()
```
